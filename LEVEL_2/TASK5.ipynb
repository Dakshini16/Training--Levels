```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df=pd.read_csv("titanic.csv")
df=df.select_dtypes(include=['number'])
corr_matrix=df.corr()
sns.heatmap(corr_matrix,annot=True,)
plt.title("heatmap")
plt.show()
```
