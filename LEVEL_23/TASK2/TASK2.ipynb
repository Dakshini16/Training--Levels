```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = "mincpcap_cppp.csv"  
df_income = pd.read_csv(file_path)
df_income_numeric = df_income.drop(columns=['country'])
avg_income = df_income_numeric.mean()
avg_income_daily = avg_income / 365
years = avg_income_daily.index.astype(int)
income_values = avg_income_daily.values
plt.figure(figsize=(10, 6))
plt.plot(years, income_values, marker='o', linestyle='-', color='g')
plt.xlabel("Year")
plt.ylabel("Average Income ($ per person per day)")
plt.title("Average Income Per Person Per Day Over Time")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("mincpcap_cppp.csv")
df_numeric = df.drop(columns=['country'])
avg_income = df_numeric.mean()
min_income = df_numeric.min()
max_income = df_numeric.max()
avg_daily = avg_income / 365
min_daily = min_income / 365
max_daily = max_income / 365
years = avg_daily.index.astype(int)
# Create the plot
plt.figure(figsize=(12, 7))
plt.plot(years, min_daily, label='Min', color='green')
plt.plot(years, max_daily, label='Max', color='brown')
plt.plot(years, avg_daily, label='Average', color='blue')
plt.title('Per Capita Income $/person/day')
plt.xlabel('Year')
plt.ylabel('Income')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
