```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = "lex.csv" 
df = pd.read_csv(file_path)
df_numeric = df.drop(columns=['country'])
avg_life_expectancy = df_numeric.mean()
years = avg_life_expectancy.index.astype(int)
life_expectancy = avg_life_expectancy.values
plt.figure(figsize=(12, 6))
plt.plot(years, life_expectancy, color='blue', linewidth=2)
plt.title("Life Expectancy Chart", fontsize=16, fontweight='bold')
plt.xlabel("Years")
plt.ylabel("Average Life Expectancy")
plt.grid(True)
plt.tight_layout()
plt.show()
```
