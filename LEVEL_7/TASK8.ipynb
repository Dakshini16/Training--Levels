```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.binomial(n=10, p=0.5, size=1000)
mean_value = np.mean(data)
plt.hist(data, bins=range(12), density=True, edgecolor='black', alpha=0.7)
plt.xlabel("Number of Successes")
plt.ylabel("Probability Density")
plt.title("Binomial Distribution (n=10, p=0.5, 1000 Experiments)")
plt.xticks(range(11))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
print(f"Mean of simulated data: {mean_value:.2f}")
print("Matches binomial expectation (n × p = 5)." if abs(mean_value - 5) < 0.5 else "Slight deviation due to randomness.")
```
