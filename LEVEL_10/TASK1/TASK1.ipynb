```python
import pandas as pd
df = pd.read_csv("titanic.csv")  
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
print("Mean Age:", df['age'].mean())
print("Median Age:", df['age'].median())
print("Mode Age:", df['age'].mode()[0])

print("Mean Fare:", df['fare'].mean())
print("Median Fare:", df['fare'].median())
print("Mode Fare:", df['fare'].mode()[0])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
categorical_features = ["sex", "pclass", "embarked"]
for feature in categorical_features:
    sns.countplot(x=feature, hue="survived", data=df)
    plt.title(f"Survival Count Based on {feature}")
    plt.show()
```

```python
import numpy as np
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
print("Survival Rate by Gender:\n", df.groupby("sex")["survived"].mean())
print("\nSurvival Rate by Class:\n", df.groupby("pclass")["survived"].mean())
```
